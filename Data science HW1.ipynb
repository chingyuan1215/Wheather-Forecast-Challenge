# balance data

```python

no_Yes = len(df[df['Attribute23'] == 'Yes'])
no_No = len(df[df['Attribute23'] == 'No'])
print(no_Yes)
print(no_No)
```

```python
Yes_indices = df[df.Attribute23 == 'Yes'].index
No_indices = df[df.Attribute23 == 'No'].index
random_indices = np.random.choice(No_indices,no_Yes, replace=False)
under_sample_indices = np.concatenate([Yes_indices,random_indices])
df = df.loc[under_sample_indices]
print(df.groupby(['Attribute23']).size())
```

```python
df=df.drop(columns=["Attribute7","Attribute6","Attribute19","Attribute18","Attribute1"],axis=1)
```

# 刪除空值

```python
# Fill missing values
categorical_features = df[['Attribute8','Attribute10','Attribute11','Attribute22','Attribute23']]
numerical_features = df.drop(['Attribute8','Attribute10','Attribute11','Attribute22','Attribute23'], axis = 1)
numerical_imputer = SimpleImputer(strategy='mean')
categorical_imputer = SimpleImputer(strategy='most_frequent')
complete_numerical = numerical_imputer.fit_transform(numerical_features)
complete_categorical = categorical_imputer.fit_transform(categorical_features)
df = pd.concat([numerical_features,categorical_features], axis=1)
```

```python
df=df.dropna(how="any")
```

```python
df.iloc[4:10]
```

```python
#轉換yes,no
df['Attribute22'].replace({'No': 0, 'Yes': 1},inplace = True)
df['Attribute23'].replace({'No': 0, 'Yes': 1},inplace = True)
```

```python
#See unique values and convert them to int using pd.getDummies()
categorical_columns = ['Attribute8', 'Attribute10', 'Attribute11']
for col in categorical_columns:
    print(np.unique(df[col]))
# transform the categorical columns
df = pd.get_dummies(df, columns=categorical_columns)
```

```python
df.iloc[4:10]
```

```python
#standardize data
from sklearn import preprocessing
scaler=preprocessing.MinMaxScaler()
scaler.fit(df)
df=pd.DataFrame(scaler.transform(df), index=df.index, columns=df.columns)
df.iloc[1:5]
```

```python
# missing values
categorical_features = [column_name for column_name in df.columns if df[column_name].dtype == 'O']
df[categorical_features].isnull().sum()
```

# outlier

```python
features_with_outliers=['Attribute3','Attribute4','Attribute5','Attribute12','Attribute13','Attribute16','Attribute17','Attribute20','Attribute21']
```

```python
for feature in features_with_outliers:
    q1 = df[feature].quantile(0.25)
    q3 = df[feature].quantile(0.75)
    IQR = q3-q1
    lower_limit = q1 - (IQR*1.5)
    upper_limit = q3 + (IQR*1.5)
    df.loc[df[feature]<lower_limit,feature] = lower_limit
    df.loc[df[feature]>upper_limit,feature] = upper_limit
```
